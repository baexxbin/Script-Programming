#### 문제 1    
시장에 장을 보러 갔습니다. 각 품목 별로 가격과 수량을 관리해서 총 지불액을 알아보려 합니다.    
사용자로부터 품목, 가격, 수량을 입력 받아 아래 실행 예시와 같이 출력하는 프로그램을 작성해 봅시다.     
(단 같은 품목을 두 번 입력할 수 없습니다. 구매품목에서 그냥 엔터만 치면 입력 종료.) 
{품목:{가격, 수량, 가격*수량}}의 형태로 딕셔너리를 만들어 사용합니다.

```python
#  빈 딕셔너리 생성
shopping = {}

sum = 0                                                # 총 지불액을 저장할 변수
while True:
    key = input("구매 품목: ")
    if key == '': break                               # 엔터입력되면 종료
    price = eval(input("가격: "))
    quantity = eval(input("수량: "))
    shopping[key] = (price, quantity, price*quantity)  # 입력받은값 딕셔너리로 저장
    sum += price*quantity
    
for key in shopping:                                   # 딕셔너리 출력
    print(key, end = ' ')
    print(str(shopping.get(key)[0]) + '원 ', end = '* ')
    print(str(shopping.get(key)[1]) + '개 ', end = '= ')
    print(str(shopping.get(key)[2]) + '원 ')
print('총 지불액: ' + str(sum) + '원')
```

#### 코드설명
1. 딕셔너리 입력받기    
먼저 빈딕셔너리를 생성한 후, key값인 구매품목과 value값인 가격과 수량을 입력받아 딕셔너리 형태로 저장한다.    
이때, price * quantity연산을 위해 value값은 eval함수를 통해 문자열이아닌 정수값으로 받아주었다.   
또한 반복문안에서 총 지불액의 sum도 연산해준다.    

2. 딕셔너리 출력    
for문을 통해 shopping 딕셔너리안의 key값과 함께 key에 해당하는 값들을 출력해준다.    
이때, 값들은 튜플형태로 저장되어있으므로 []의 인덱스 연산을 이용한다.    

#### 문제 2    
A매장과 B매장이 있습니다. 5개의 튜플을 원소로 갖는 리스트에 두 매장의 5일치 요일 별 매출을 기록한 데이터가 있다고 가정합시다. 하나의 튜플은 두 매장의 매출액을 A, B 순서로 나타냅니다. 각 매장의 매출이 전날보다 감소한 날이 총 며칠이고 무슨 요일에 감소했는지 출력하는 프로그램을 작성해 봅시다. (순서대로 월요일부터 금요일로 가정)

```python
# 값,변수 초기화
sales = [(100, 90), (115,100), (85, 103), (64,105), (124,60)]
count = [0,0]
day = [[],[]]

# 전날대비 매출 감소일 판별
for i in range(0, len(sales)):
    if i+1 < len(sales):                  # i+1이 len(sales)를 넘지않을때
        if sales[i+1][0] < sales[i][0]:   # A매장 전날대비 매출감소
            count[0] += 1
            day[0].append(i+1)
        if sales[i+1][1] < sales[i][1]:   # B매장 전날대비 매출감소
            count[1] += 1
            day[1].append(i+1)
# 값 출력
print("A매장 전날 대비 매출 감소일: " + str(count[0]) + "일")
print("B매장 전날 대비 매출 감소일: " + str(count[1]) + "일")

print("A매장 전날 대비 매출 감소요일: ", end='')
for x in range(len(day[0])):
    if day[0][x] == 0: print('월요일', end=' ')
    elif day[0][x] == 1: print('화요일', end=' ')
    elif day[0][x] == 2: print('수요일', end=' ')
    elif day[0][x] == 3: print('목요일', end=' ')
    elif day[0][x] == 4: print('금요일', end=' ')
print()
print("B매장 전날 대비 매출 감소요일: ", end='')
for x in range(len(day[1])):
    if day[1][x] == 0: print('월요일', end=' ')
    elif day[1][x] == 1: print('화요일', end=' ')
    elif day[1][x] == 2: print('수요일', end=' ')
    elif day[1][x] == 3: print('목요일', end=' ')
    elif day[1][x] == 4: print('금요일', end=' ')
```

#### 코드설명
1. 값, 변수 초기화    
sales의 튜플형태의 값과, 전날 대비매출 감소 일과 요일을 저장할 변수 count와 day를 배열형태로 초기화해준다.    

2. 전날대비 매출 감소일 판별    
for문을 돌며 전날보다 매출이 감소한 날을 count를 통해 세고, 해당일을 day에 저장한다. 이때, i가 리스트의 범위가 넘어가지 않게하기위해 i+1 < len(sales)라는 조건을 걸어준다.    

3. 값 출력    
배열형태로 저장된 값들을 출력해준다.    
A매장은[0], B매장은[1]에 해당한다. 해당요일은 if문을 통해 숫자별 요일을 알도록 하였다.    

#### 문제 3    
세트와 반복문을 사용하여 주사위 3개를 던져서 얻을 수 있는 모든 경우의 수를 출력하는 프로그램을 작성합니다.

```python
# 세트와 튜플 선언
case = set()

# 3중반복문을 돌면서 경우의 수에 해당하는 값 생성
for i in range(1,7):
    for j in range(1,7):
        for k in range(1,7):
            tuple1 = (i, j, k)
            case.add(tuple1)
            
# 결과 출력
print(case)
```

#### 코드설명
1. 세트와 튜플선언      
각 경우의 수를 저장할 튜플과 그 튜플들을 저장할 세트를 선언한다.    

2. 3중반복문을 통해 경우의 수 생성    
주사위 3개를 던지므로 3중for문을 이용한다.    
튜플에 (i,j,k)의 값을 저장하고 add연산을 통해 세트에 추가한다.    

3. 결과출력    
저장된 세트를 출력한다.    

#### 문제 4    
다음과 같이 주어진 키와 값의 쌍으로 이루어진 fruits_dic이라는 이름의 딕셔너리를 생성합니다. fruits_dic 딕셔너리의 모든 키와 값을 각각 따로 출력하고 특정 과일이 있는지 없는지 검사하여 출력하는 프로그램을 작성합니다. (과일 이름은 입력을 받습니다. 그냥 엔터만 치면 프로그램 종료.)

```python
# 딕셔너리 생성
fruits_dic = {"apple":"6000", "melon":"3000", "banana":"5000", "orange":"4000"}

# 딕셔너리 키,값 출력
print("딕셔너리의 키는")
print(list(fruits_dic.keys()))
print()
print("딕셔너리의 값은")
print(list(fruits_dic.values()))

# 과일 검사
while True:
    key = input("검사하고 싶은 과일은? ")
    if key == '':
        print("종료합니다.")
        break
    
    if fruits_dic.get(key) == None: print("해당 과일은 없습니다.")
    else: print("해당 과일은 있습니다.")
```

#### 코드설명
1. 딕셔너리 생성     
과일과 값을 딕셔너리로 생성한다.    

2. 딕셔너리 키,값 출력    
.keys(), .values()매소드를 이용하여 딕셔너리의 키와 값을 각각 출력한다.    

3. 과일검사    
무한반복문을 돌면서 입력으로 엔터를 받았다면 '종료합니다'를 출력하며 종료한다.    
if문을 통해 검사하고자하는 값을 .get()매소드를 이용해 검사한다.    
값이 없을 시 None을 가짐으로 이를 기준으로 if문을 실행한다.
