#### 문제 1
n * n 2차원 리스트를 선언합니다. n은 3~10 사이의 값으로 사용자에게 입력 받습니다.   
리스트의 원소는 0 또는 1로 랜덤하게 초기화합니다.   
리스트의 모든 원소를 출력한 후 리스트 내에서 0인 값을 -1로 바꿔주고 이를 다시 출력합니다.

```python
import random

# 리스트 선언
matrix = []

# 사용자로부터 n값 입력받기
n = eval(input("3~10사이의 n의 값을 입력하세요: "))

# n*n리스트를 0 또는 1로 초기화
matrix = [[random.randint(0,1) for col in range(n)] for row in range(n)]

# 리스트 이쁘게 출력
for row in range(0,len(matrix)):
    for col in range(0, len(matrix[row])):
        print(matrix[row][col], end= "\t")
    print()
    
print()
# 0을 -1로 바꾸기
for row in range(n):
    for col in range(n):
        if matrix[row][col]== 0:
            matrix[row][col] = -1
        
# 리스트 이쁘게 출력
for row in range(len(matrix)):
    for col in range(len(matrix[row])):
        print(matrix[row][col], end= "\t")
    print()
```

#### 코드설명
1. 2차원 리스트를 위한 리스트를 선언한다.

2. 사용자로부터 n값을 입력받는다.

3. 리스트 컴프리헨션을 통해 2차원리스트를 랜덤값으로 초기화한다.    
첫번째 for문은 col, 두번째 for문은 row가 된다.

4. 이중for문을 통해 값 바꾸기     
matrix[row][col]의 값이 0일때 -1로 바꾼다. 

#### 문제 2
n * n 2차원 리스트를 생성합니다.n은 3~10 사이의 값으로 사용자에게 입력 받습니다.   
리스트가 다음과 같이 되도록 만드시오. 

실행 예시 n = 3인 경우    
1  2  3    
6  5  4    
7  8  9    
 
실행 예시 n = 4인 경우    
1  2  3  4    
8  7  6  5    
9  10 11 12    
16 15 14 13    

위와 같이 리스트가 출력되어야 합니다.   

```python
# n*n 리스트 생성
matrix = []
n = eval(input("3~10 사이의 n값을 입력하시오: "))

# 리스트 0으로 초기화
matrix = [[0 for col in range(n)] for row in range(n)]

# ㄹ자로 숫자출력
num = 1
for row in range(n):
    ne = -1
    for col in range(n):
        if row%2 == 0: matrix[row][col] = num
        else: matrix[row][ne] = num
        num+=1
        ne -=1

# 리스트 출력
for row in range(0,len(matrix)):
    for col in range(0, len(matrix[row])):
        print(matrix[row][col], end= "\t")
    print()       
    
# print(matrix[2])
# for i in range(len(matrix[2])):
#     if matrix[2][i] == 11:
#         print(i)
        
# print(matrix[2][i] == 11)
```

#### 코드설명
1. n*n리스트 생성    
matrix 리스트를 선언하고, 사용자로부터 n값을 입력받는다.

2. 리스트를 0으로 초기화    
숫자를 넣을때 편하기위해 미리 이차원리스트를 0으로 초기화 해놓았다.    

3. ㄹ자로 숫자출력    
ㄹ자로 숫자를 오름차순으로 출력하기 위해서는 짝수 행에는 순서대로, 홀수 행에는 반대로 출력이되어야한다.    
따라서 짝수행일때는 순서대로 출력을 하고, 홀수행일경우 -1로 초기화된 ne변수를 이용해 뒤에서부터 출력하도록 하였다.   

4. 리스트 출력    
완선된 리스트를 출력한다.

#### 문제 3
두 개의 10 * 10 크기의 2차원 리스트를 선언합니다. 리스트 초기 값은 1~9 사이의 정수 값으로 랜덤하게 초기화 합니다.    
두 개의 리스트를 A, B라고 했을 때, 두 리스트의 각 원소를 더합니다. 더한 값이 10 미만이면 그대로, 더한 값이 10 이상이면 10을 빼고 이를 출력합니다.

```python
import random

# 2개의 2차원리스트 선언
A = []
B = []

# 리스트 랜덤값으로 초기화
A=[[random.randint(1,9) for col in range(10)]for row in range(10)]
B=[[random.randint(1,9) for col in range(10)]for row in range(10)]

# 두 리스트를 합한 값을 저장할 리스트 0으로 초기화
add=[[0 for col in range(10)]for row in range(10)]

# 두 리스트 더하기
for row in range(10):
    for col in range(10):
        add[row][col] = A[row][col]+B[row][col]
        if add[row][col] >= 10:
            add[row][col]-=10

# 리스트 출력
for row in range(10):
    for col in range(10):
        print(A[row][col], end= " ")
    print()    
    
print('+')
        
for row in range(10):
    for col in range(10):
        print(B[row][col], end= " ")
    print()         

print('=')    
        
for row in range(10):
    for col in range(10):
        print(add[row][col], end= " ")
    print()        
```

#### 코드설명
1. 2개의 2차원리스트 선언

2. A,B리스트 랜덤값으로 초기화    
random함수를 이용하여 A,B리스트의 값을 1~9까지의 수로 초기화한다.     

3. add리스트 생성과 초기화    
A,B리스트를 합한값을 저장할 리스트를 길이10의 0으로 초기화한다.    

4. A,B리스트 더하기    
이중for문을 돌면서 A,B리스트의 값을 더해 add리스트에 넣어준다.    
이때, 더한값이 10이상이면 10을 빼준다.    

5. 리스트 출력
A,B,add 리스트를 보기좋게 출력한다.

#### 문제 4
빈도수 세기, 대문자로 변경 하기.    
test_1.txt 파일 중 savage라는 단어가 몇 번 등장하는지 출력하고, savage를 Savage로 변경하여 파일 내용을 출력하는 프로그램을 작성하시오.
*test_1.txt는 따로 제공합니다.    

```python
# 파일 열기
filename = input("파일명을 입력하세요: ").strip()
infile = open(filename, 'r', encoding='UTF8')

# savage 개수세기 & Savage로 변경하여 출력
count = 0
for lines in infile:
    for ch in lines.split():
        lines = lines.replace('savage', 'Savage')
        if ch == 'savage':count += 1
    print(lines)

# 총 savage 수 출력
print(f'savage는 {count}번 등장합니다.')
```

#### 코드설명
1. 파일열기    
읽기모드로 파일을 연다.    
한글파일임으로 encoding='UTF8'을 추가해준다.    

2. savage 개수세기 & Savage로 바꿔 출력하기
lines로 파일을 읽고, lines안에 ch로 단어를 읽으면서, replace()를 이용하여 savage를 Savage로 바꾸어준다.      
ch가 savage일때 count ++를 해준다.    

3. 총 savage수 출력    
count를 출력해준다.   

#### 문제 5
(줄 번호 매기기)    
test_1.txt 파일을 읽어서 각 줄마다 번호를 출력하고 2칸 띄워서 원래 텍스트를 출력하는 프로그램을 작성하시오. 

```python
# 파일 열기
filename = input("파일명을 입력하세요: ").strip()       # 파일이름 입력받기
infile = open(filename, 'r', encoding='UTF8')           # 파일 객체 생성

# 번호와함께 텍스트 출력
n = 1
for lines in infile:
    print(n, lines, sep='  ')
    n += 1
```

#### 코드설명
1. 파일열기

2. 번호와 함께 텍스트 출력    
번호를 매길변수 n을 설정한 후, 한문장이 끝날때마다 ++해준다.    
번호와 텍스트사이에 두칸을 띄우기 위해 sep='  '을 사용하였다.
